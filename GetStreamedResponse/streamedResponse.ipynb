```python
import semantic_kernel as sk
from dotenv import dotenv_values

config = dotenv_values()
kernel = sk.Kernel()
```

```python
from semantic_kernel.connectors.ai.open_ai import AzureChatCompletion

azure_completion_service = AzureChatCompletion(
        service_id="streamed_output",
        deployment_name=config["AZURE_OPENAI_DEPLOYMENT"],
        api_key=config["AZURE_OPENAI_API_KEY"],
        api_version=config["AZURE_API_VERSION"],
        endpoint=config["AZURE_OPENAI_ENDPOINT"]
    )
```

```python
from semantic_kernel.connectors.ai.open_ai.prompt_execution_settings.azure_chat_prompt_execution_settings import AzureChatPromptExecutionSettings

streamed_output_execution_settings = AzureChatPromptExecutionSettings(
    service_id="streamed_output",
    max_tokens=150
)
```

```python
from semantic_kernel.contents.chat_history import ChatHistory

chat = ChatHistory()

chat.add_system_message("You are an AI assistand that can create poems about apricots.")
chat.add_user_message("Tell me a poem on a rainy day.")

stream = azure_completion_service.complete_chat_stream(
    chat_history=chat,
    settings=streamed_output_execution_settings
)

async for text in stream:
    print(str(text[0]), end="")
```
